# Covid19 Bar Chart Race Graph

# Import all related libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

# Load it into DataFrame

```python
df = pd.read_csv('Covid_19_Data.csv')
df.head(10)
```

# Sort the DataFrame on Country and State

```python
df = df.sort_values(by=['Province/State','Country/Region'])
df.head(10)
```

```python
df=df.reset_index(drop=True)
```

# Unpivot the table

```python
dates_df=df.columns[4:]
unpivot_df = df.melt(id_vars=['Province/State','Country/Region','Lat','Long'], value_vars=dates_df,var_name='Date', value_name = "Confirmed")
unpivot_df.head()
```

# Converting the Date Column to Date Format

```python
unpivot_df['Date'] = unpivot_df['Date'].apply(lambda x : datetime.datetime.strptime(x, '%m/%d/%y').date())
unpivot_df.head()
```

# Grouping By Date and Country/Region and Summing up the Confirmed Cases for Each Country

```python
df_daily = unpivot_df.groupby(['Date','Country/Region']).sum()
df_daily
```

# Sorting the DataFrame by Date and Confirmed

#We want to sort the df_dailyDataFrame based on two keys:
#Date (ascending order)
#Confirmed (descending order)
#So the following statements accomplish that:

```python
df_daily_sorted = df_daily.sort_values('Confirmed',ascending=False).sort_index(level=0,ascending = True)
df_daily_sorted
```
